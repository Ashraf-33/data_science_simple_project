In this project i try to find those people who survied.<br>
By using csv file of titanic data.

```python
#First import useful package 
# And cread dataframe to csv file
import pandas as pd
import numpy as np
# read_csv
url = r"C:\Users\ASHRAF\Downloads\titanic\train.csv"
df = pd.read_csv(url)
df.head(10)
```

```python
# first copy this data frame:
df_train = df.copy()
```

<h1>1. first find missing values and get the idea about this dataframe.</h1>

```python
print("the shape of this df:\n")
print(df_train.shape)
print("\n the info of data:\n")
print(df_train.info())
# find missing values:
print("\n is there any missing value availabe:\n",df_train.isnull().sum())
```

<p>So i found 3 column which have lots of missing value.<br>
the <b>Cabin</b> column have majority number of missing values and i think it doesnot useful now.<br>
For Age and Embarked i will creat new column where the missing value fill with their median and mode.</p>

```python
#For Age and Embarked column:
df_train['Age_Filled']=df_train['Age'].fillna(df_train['Age'].median())
df_train['Embarked_Filled']=df_train['Embarked'].fillna(df_train['Embarked'].mode())
df_train.head(10)
```

```python
# drop the cabin column:
df_train.drop('Cabin',axis=1,inplace=True)
df_train.head()
```

so as their gender(Sex) and Embarked string data.<br>
for ml it could be an obsticle so remove it by numerical data.

```python
# convert male and female by 0 and 1:
df_train['Sex']=df_train['Sex'].map({'male':0,'female':1})
# convert C,Q,S by 0,1,2:
df_train['Embarked_Filled']=df_train['Embarked_Filled'].map({'C':0,'Q':1,'S':2})
print(df_train.head())
```

<h1>Explore Data(EDA)</h1>

```python
# show the data of surviver by age:
# Note in survivour section 0=no and 1=yes.
print(df_train[df_train['Survived']==1]['Age'].describe())
```
