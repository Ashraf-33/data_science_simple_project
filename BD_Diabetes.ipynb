```python
# lets find the relation between gulcose and bmi:
# for this i will use scatter plot:
sns.scatterplot(data=data_df,x='bmi',y='glucose',hue='diabetic',alpha=0.7)
plt.show()
```

Well there some problem i found you can see some of the value of bmi indicates more than 100.<br>
Which is totally unrealistic so first should remove the outlier and then see what happend.

```python
#Remove outlier:
# i do not want to modify the data so i will do this on copy data;
copy_data = data_df.copy()
copy_data=copy_data[copy_data['bmi']<50]
```

```python
# lets visualize :
sns.scatterplot(data=copy_data,x='bmi',y='glucose',hue='diabetic',alpha=0.7)
#give axis limit:
plt.xlim(0,60)
plt.ylim(0,40)

plt.title('Scatter BMI Vs Glucose',fontsize=15)
plt.xlabel('BMI')
plt.ylabel('Glucose')
plt.show()
```

<b>Confession:</b> it tells majotrity of people those have 10 to 30 bmi lavel and 0 to 10 lavel of glucose are not 
in diabetic.<br>
But those have avobe 15 in glucose are mostly faced diabetic though they have 10 to 30 bmi lavel.<br>
<b> So it tells diabetic is habily related with glucose lavel.</b><br>
We should control our suger(glucose) lavel bellow 15.<br>
Lastly, we can see some people have diabetic but their glucose lavel is normal. so it indicate there factor which is cause of diabetc.<br>
On the other hand some got diabetic with high bmi.

```python
# Extra :
# find peaple who are obese or not.
# i will use bmi for find it :
def body_cat(bmi):
    if bmi <= 19:
        return 'underweight'
    elif bmi <=26:
        return 'normal'
    elif bmi > 26:
        return 'overweight'
        
copy_data['Body_category']=copy_data['bmi'].apply(body_cat)
```

```python
copy_data.head(10)
```

```python
# now visualize between Body_category and diabetic:
sns.countplot(data=copy_data,x='Body_category',hue='diabetic')
plt.title('Body_category by diabetic')
plt.show()
```

<b>Confession:</b> so we can see though there some people who have diabetic whose are in underweight or overweight.<br>
But more people are in diabetic while they are normal in body category.<br>
So body category or bmi doesnot much involve in diabetic.
