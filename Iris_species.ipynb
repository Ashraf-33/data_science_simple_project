```python
# import important liberaries:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# first copy the path of Iris in url:
url=r"C:\Users\ASHRAF\Downloads\Iris.csv"
# read and make a dataframe of this first:
iris_data = pd.read_csv(url)
# show the data
iris_data.head()
```

```python
# Before going to deep let us analyze the data :
print(iris_data.info())
```

<b>conclusion:</b> so this information tells this data have 150 row and 6 columns and other information.<br>

```python
# use isnull: and use sum for getting this in total
print(iris_data.isnull().sum())
```

so, there are used 3 types of species for 50 times.

<h1>Visualize the data for understand:</h1>

```python
# use scatter plot for find the realationship between SepalLengthCm and SepalWidthCm.
x = iris_data['SepalLengthCm']
y = iris_data['SepalWidthCm']
plt.scatter(x,y)
plt.title('Scatter of SepalLengthCm And SepalWidthCm')
plt.xlabel('Sepal_Length_Cm')
plt.ylabel('Sepal_Width_Cm')
plt.show()
```

```python
# ok though it shows scatter, for better understanding
#i use seaborn and categories the species:
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',hue='Species',data=iris_data)
plt.title("The Scatter of SepalLengthCm and SepalWidthCm by species",fontsize=15)
plt.xlabel('Sepal_Length_(Cm)',fontsize=15)
plt.ylabel('Sepal_Width_(Cm)',fontsize=15)
plt.show()
```

<p>ok this scatter tells <b>Iris-versicolor and Iris_virginica</b> both of the species
almost have similar Sepal lenght and width though virginica shows some outlier.<br>
But<b> Iris-setosa</b> is totally different from them in terms of length and width.</p>

```python
df=iris_data.copy() # i don't want to risk my real data
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(10,8))
sns.boxplot(x='Species',y='PetalLengthCm',data=df,label='Petal_Length',color='r')
sns.boxplot(x='Species',y='PetalWidthCm',data=df,label='Patel_Width',color='g')
plt.title('Boxplot by Petal_length_and_width')
plt.xlabel('Species',fontsize=13,color='k')
plt.ylabel('Petal_length_and_width',fontsize=13,color='k')
plt.legend(fontsize=15)
plt.show()
```

So this boxplot describe the statistical information of petal length and width by there
species.

```python
# let's use pairplot to get relationship of all columns.
# first drop the Id columns there are not work of it
sns.pairplot(df.drop('Id',axis=1),hue='Species',height=3)
plt.show()
```

So, most of the cases Iris_setosa species is unique than rest of the species.

```python
#lastly make a heat map of this.
df=df.drop(columns=['Id','Species'])
sns.heatmap(df.corr(),annot=True,cmap='rocket_r')
plt.show()
```
